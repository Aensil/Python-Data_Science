```python
def evalua_vocal(valor):
  if valor is "a" or valor is "e" or valor is "i" or valor is "o" or valor is "u" or valor is "A" or valor is "E" or valor is "I" or valor is "O" or valor is "U":
    return True
  else:
    return False

def evalua_numero(valor):
  if valor is "1" or valor is "2" or valor is "3" or valor is "4" or valor is "5" or valor is "6" or valor is "7" or valor is "8" or valor is "9" or valor is "0":
    return True
  else:
    return False

def evalua_caracter(valor):
  if valor is "|" or valor is "!" or valor is "#" or valor is "$" or valor is "%" or valor is "&" or valor is "/" or valor is "(" or valor is ")" or valor is "=" or valor is "'" or valor is "?" or valor is "¿" or valor is "´´" or valor is "+" or valor is "{" or valor is "}" or valor is ","or valor is "." or valor is "*" or valor is ";":
    return True
  else:
    return False

valor=input("escriba un caracter ---> ")
b=evalua_vocal(valor)
c=evalua_numero(valor)
d=evalua_caracter(valor)
if b==True:
  print("Es una Vocal: ", valor)
elif c==True:
  print("Es un Numero: ", valor)
elif d==True:
  print("Es un Caracter Especial: ", valor)
else:
  print("Es un Caracter: ", valor)
```

```python
from random import seed
from random import randint


def punto_uno():
  try:
    b=0
    asig=[]
    nota=[]
    while b==0:
      c=input("¿Desea agregar una asignatura? \n 1 para si \n otro numero para no \n")
      c=int(c)
      if c==1:
        d=input("Ingrese la asignatura: ")
        print(d)
        asig.append(d)
        e=input("Ingrese la nota: ")
        nota.append(e)
      else:
        print("--------------------------------------------------------------")
        for j in range(len(asig)):
          print("{0} nota: {1}".format(
            asig[j], nota[j]))
        print("------------------------------------------------------------")
        break
  except:
    print("Adios")
    print("---------------------------------------------------------")


def punto_dos():
  order=[]
  non=[]
  for j in range(10):
    f = randint(100, 999)
    non.append(f)
    order.append(f)
  order.sort()
  print("------------------------------------------------------------")
  print("Lista en desorden: ",non)
  print("Lista en orden: ", order)
  print("------------------------------------------------------------")

def punto_tres():
  order=[]
  non=[]
  num=[1,2,3,4,5,6,7]
  for j in range(7):
    f = randint(1, 100)
    non.append(f)
    order.append(f)
  order.sort()
  print("------------------------------------------------------------")
  for j in range(7):
    print("--------")
    print("Producto {0} precio: {1}".format(
            num[j], order[j]))
  print("------------------------------------------------------------")

def punto_cuatro():
  g=input("Write the phrase: \n")
  g=str(g)
  h=g.split()
  k=[]
  for j in range(len(h)):
    if len(h[j]) > 5:
            k.append(h[j])
  l=len(k)
  print("------------------------------------------------------------")
  print("La cadena tiene ", l, "palabras de mas 5 caracteres")
  print("La cadena es: \n",k)
  print("------------------------------------------------------------")

def punto_cinco():
  g=input("Write the message: \n")
  g=str(g)
  k=[]
  l=''
  for j in range(len(g)):
    if g[j] == '.':
      l= l + ' STOP '
    else:
      l=l + g[j]
  h=l.split()
  o=''
  z=0
  for p in range(len(h)):
    if len(h[p]) > 5:
      o=h[p][0:5]
      k.append(o+'@')
      z+=1
    else:
      k.append(h[p])
  
  s=''
  for q in range(len(k)):
    for r in range(len(k[q])):
      s= s + k[q][r]
      if r == len(k[q])-1:
        s= s + ' '
  s= s + 'STOPSTOP'
  print("------------------------------------------------------------")
  print("la frase a comunicar es: ",s)
  print("el numero de palabras cortadas es: ",z)
  print("------------------------------------------------------------")

def punto_seis():
  t=list(range(0,4))
  u=list()
  nn=0
  result=[]
  for j in range(len(t)):
    xx=2**(t[j]+1)
    u.append(xx)
  for z in range(len(t)+1):
    for m in range(z):
      nn=nn + (1/u[m])
      if m == z-1:
        result.append(nn)
        nn=0
  print("------------------------------------------------------------")     
  print("Result: ",result)
  print("------------------------------------------------------------")


n=0
while n==0:
  try:
    a=input("Por favor ingrese el numero 1-6 para seleccionar el ejercicio: \n 0 para salir \n")
    a=int(a)
    if a==1:
      punto_uno()
    elif a==2:
      punto_dos()
    elif a==3:
      punto_tres()
    elif a==4:
      punto_cuatro()
    elif a==5:
      punto_cinco()
    elif a==6:
      punto_seis()
    elif a==0:
      break
    else:
      print("Ingrese un numero valido")
  except:
    print("Ingrese un valor valido")
```

```python
import pandas as pd    #pandas
import numpy as np    #numpy

import matplotlib.pyplot as plt  #para graficar matplotlib

df_Tunja=pd.read_csv("TUNJA.csv", encoding='latin-1')#leer de archivo en colab
df_Cartagena=pd.read_csv("CARTAGENA DE INDIAS.csv", encoding='latin-1')#leer de archivo en colab
df_Charala=pd.read_csv("CHARALÁ.csv", encoding='latin-1')#leer de archivo en colab
df_Ipiales=pd.read_csv("IPIALES.csv", encoding='latin-1')#leer de archivo en colab

print('-------------------------------------------------------------------')
df_all = pd.concat([df_Tunja, df_Cartagena, df_Charala, df_Ipiales])
print(df_all)
print('-------------------------------------------------------------------')
print('-------------------------------------------------------------------')
promedio_Tunja = df_all[df_all.Departamento == 'BOYACÁ']['Temperatura'].mean()
print('La temperatura promedio de tunja: ', promedio_Tunja)
promedio_Ipiales = df_all[df_all.Departamento == 'NARIÑO']['Temperatura'].mean()
print('La temperatura promedio de Ipiales: ', promedio_Ipiales)
promedio_Cartagena = df_all[df_all.Departamento == 'BOLÍVAR']['Temperatura'].mean()
print('La temperatura promedio de Cartagena: ', promedio_Cartagena)
promedio_Charala = df_all[df_all.Departamento == 'SANTANDER']['Temperatura'].mean()
print('La temperatura promedio de Charala: ', promedio_Charala)
print('-------------------------------------------------------------------')
print('-------------------------------------------------------------------')
max_Tunja = df_all[df_all.Departamento == 'BOYACÁ']['Temperatura'].max()
print('Valor maximo de tunja: ', max_Tunja)
print('-------------------------------------------------------------------')
print('-------------------------------------------------------------------')
tunja=df_all[df_all.Departamento == 'BOYACÁ']
print('Tunja tiene: ', len(tunja), ' Registos')
Ipiales=df_all[df_all.Departamento == 'NARIÑO']
print('Ipiales tiene: ', len(Ipiales), ' Registos')

print('-------------------------------------------------------------------')
print('-------------------------------------------------------------------')
tunja_1 = df_all[df_all.Departamento == 'BOYACÁ']['Temperatura']
plt.plot(tunja_1, label='Tunja')
Ipiales_1 = df_all[df_all.Departamento == 'NARIÑO']['Temperatura']
plt.plot(Ipiales_1, label='Ipiales')
Cartagena_1 = df_all[df_all.Departamento == 'BOLÍVAR']['Temperatura']
plt.plot(Cartagena_1, label='Cartagena')
Charala_1 = df_all[df_all.Departamento == 'SANTANDER']['Temperatura']
plt.plot(Charala_1, label='Charala')
plt.title("Temperaturas")
plt.ylabel("°C")
plt.xlabel("fechas")

plt.legend() #imprime labels

plt.show()
print('-------------------------------------------------------------------')
print('-------------------------------------------------------------------')
df_all.reset_index(drop=True, inplace=True)   #resetea el indix
print('-------------------------------------------------------------------')
print('-------------------------------------------------------------------')
for index in df_all.index:
    if df_all.loc[index,'Departamento']=='BOYACÁ':
        df_all.loc[index,'Región'] = 'Tunja'
    elif df_all.loc[index,'Departamento']=='NARIÑO':
        df_all.loc[index,'Región'] = 'Ipiales'
    elif df_all.loc[index,'Departamento']=='BOLÍVAR':
        df_all.loc[index,'Región'] = 'Cartagena'
    else:
        df_all.loc[index,'Región'] = 'Charala'


print('-------------------------------------------------------------------')
print('-------------------------------------------------------------------')

df_all.to_csv('all.csv') #exporta el dataframe
```
